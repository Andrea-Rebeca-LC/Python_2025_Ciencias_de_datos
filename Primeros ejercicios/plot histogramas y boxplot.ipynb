Analizar base y usar pandas

```python
import pandas as pd

# Import the air quality dataset 
CO2Data = pd.read_csv("Terminos_lagoon_TA_DIC_2023_RawData.csv")


print(CO2Data.info())
```

Vamos a crear un un plot de un histograma 

```python
import matplotlib.pyplot as plt

# Select data
sal_data = CO2Data["sal_psu"]

plt.hist(sal_data, bins=5)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")
plt.show()
```

```python
#histograma por densidad 
data = CO2Data["sal_psu"]


plt.hist(data, bins=10, density=True)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")
plt.show()
```

boxplot

```python
do_data = CO2Data["do_mg_l"]
plt.boxplot(do_data)

plt.xlabel("DO (mg/L)")
plt.ylabel("Concentration")
plt.title("Boxplot of DO Concentration")

# show plot
plt.show()
```

```python
# boxplot para salinidad con otro formato
data = CO2Data["sal_psu"]

fig2, ax2 = plt.subplots()

ax2.set_title('Notched boxes')
ax2.boxplot(data, notch=True)

plt.ylabel('Salinity ($PSU$)', fontsize = 12)
plt.show()
```

```python
#  boxplot para TA
green_diamond = dict(markerfacecolor='g', marker='D')
fig3, ax3 = plt.subplots()

ax3.set_title('Changed Outlier Symbols')

ax3.boxplot(data, flierprops=green_diamond)


plt.ylabel('TA ($\mu mol  \; kg^{-1}$)', fontsize = 12)
plt.show()
```

Tarea hacer plot con histogramas 

```python
import matplotlib.pyplot as plt

# datos de dic y ta
ta_data = CO2Data["ta_micromol_kg"]
dic_data = CO2Data["dic_micromol_kg"]

# Crear histograma para TA y DIC
plt.hist([ta_data, dic_data], bins=10, label=["TA", "DIC"], alpha=0.4)  # alpha controla la transparencia

# Etiquetas y título
plt.xlabel("Concentration (µmol/kg)")
plt.ylabel("Frequency")
plt.title("Histogram of TA and DIC")
plt.legend()  # muestra la leyenda

# Mostrar
plt.show()
```
