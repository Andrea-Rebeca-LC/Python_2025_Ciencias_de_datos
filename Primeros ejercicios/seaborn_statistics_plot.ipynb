Visualizaci√≥n estadistica con Seaborn

```python
import pandas as pd
# Import the air quality dataset 

CO2Data = pd.read_csv("Terminos_lagoon_TA_DIC_2023_RawData.csv")

print(CO2Data.info())
```

se crean las librerias y figuras 

```python
import seaborn as sns
import matplotlib.pyplot as plt 

#create your own color array (viridis)
my_colors = ["#5ec962", "#3b528b"]

# add color array to set_palette
sns.set(font_scale=2, style="ticks", palette= my_colors)
```

diagrama de caja 

```python
ax = sns.catplot(data=CO2Data,
                 x="area", y="ta_micromol_kg",
                hue="season", col="estuary",
                kind="box", 
                height=7, aspect=1)

  
ax.set(xlabel='', ylabel=' TA ($\mu mol \; kg^{-1}$)')

ax.set_titles(col_template="{col_name}", row_template="{row_name}")

plt.setp(ax._legend.get_title(), fontsize=20)

sns.despine(top=False,right=False)

plt.show()
```

violin y catplot

```python
ax = sns.catplot(data=CO2Data, 
                x="area", y="ta_micromol_kg",
                hue="season", col="estuary", 
                kind="violin", 
                height=7, aspect=1)

  
ax.set(xlabel='', ylabel=' TA ($\mu mol \; kg^{-1}$)')

ax.set_titles(col_template="{col_name}", row_template="{row_name}")

plt.setp(ax._legend.get_title(), fontsize=20)

sns.despine(top=False,right=False)

plt.show()
```

ANOVA

```python
# grafico de puntos anova de 3 v√≠as
ax = sns.catplot(
    data=CO2Data, x="season", 
    y="ta_micromol_kg", hue="estuary", col="area",
    capsize=.2, palette="YlGnBu_d", errorbar="se",
    kind="point", height=6, aspect=.75,
)
ax.despine(left=True)

ax.set(xlabel='', ylabel=' TA ($\mu mol \; kg^{-1}$)')

ax.set_titles(col_template="{col_name}", row_template="{row_name}")

plt.setp(ax._legend.get_title(), fontsize=20)

sns.despine(top=False,right=False)

plt.show()
```

diagramas de dispesi√≥n 

```python
ax = sns.lmplot(x="sal_psu", y="ta_micromol_kg",
                hue="season",col="estuary",
                data=CO2Data, 
                height=7, aspect=1)

ax.set(xlabel='Salinity (PSU)', ylabel=' DIC ($\mu mol \; kg^{-1}$)')

ax.set_titles(col_template="{col_name}", row_template="{row_name}")

plt.setp(ax._legend.get_title(), fontsize=20)

sns.despine(top=False,right=False)

plt.show()
```

```python
import pandas as pd
# Import the air quality dataset 

CHL = pd.read_csv("insitu y satelital.csv")

print(CHL.info())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

#paleta de colores
my_colors_E = ["#74add1", "#4575b4", "#313695","#006837","#a6d96a",  "#fdae61","#f46d43","#d73027","#a50026", "#67000d"]

# add color array to set_palette
sns.set(font_scale=2, style="ticks", palette= my_colors_E)
```

Grafico de dispersi√≥n 

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

ax = sns.scatterplot(
    data=CHL,
    x="chlorophyll.ugl",
    y="mdn",
    hue="station",
    palette=my_colors_E,
    legend="full"
)

# Personalizar leyenda
plt.legend(
    bbox_to_anchor=(1.05, 1),   # a mueve fuera a la derecha
    loc='upper left',           #  anclada arriba a la izquierda del cuadro
    borderaxespad=0,
    title="Estaci√≥n",
    prop={'size': 8}            #  tama√±o de texto 
)

plt.xlabel(r"EXO Chl-$\it{a}$  ($\mu g \, L^{-1}$)", fontsize=12)
plt.ylabel(r"Sentinel-2 MDN Chl-$\it{a}$  ($\mu g \, L^{-1}$)", fontsize=12)

plt.tight_layout()
plt.show()
```

Bloxplot por estaci√≥n 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# organizar datos 
df_long = CHL.melt(
    id_vars="station",                          # columna que se mantiene
    value_vars=["oc3", "ocx", "mdn", "chlorophyll.ugl"],  # columnas a comparar
    var_name="Variable", 
    value_name="Concentraci√≥n"
)

plt.figure(figsize=(12,6))

sns.boxplot(
    data=df_long,
    x="station",         # estaciones en eje X
    y="Concentraci√≥n",   # valores en eje Y
    hue="Variable",      # cada variable con distinto color
    palette="Set2"       # cambiar los colores
)

plt.xlabel("Estaci√≥n", fontsize=12)
plt.ylabel("Concentraci√≥n de Clorofila (¬µg/L)", fontsize=12)
plt.title("Comparaci√≥n de m√©todos de clorofila por estaci√≥n", fontsize=14)

plt.legend(title="M√©todo", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Renombrar variables
df_long_renamed = df_long.copy()

# Nuevos nombres
rename_dict = {
    "oc3": "OC3 (Sentinel-2)",
    "ocx": "OCx (Sentinel-2)",
    "mdn": "MDN (Sentinel-2)",
    "chlorophyll.ugl": "EXO Chl-a (in situ)"
}

# Reemplazar los nombres
df_long_renamed["Variable"] = df_long_renamed["Variable"].map(rename_dict)

# Boxplot
plt.figure(figsize=(12,6))

sns.boxplot(
    data=df_long_renamed,
    x="station",
    y="Concentraci√≥n",
    hue="Variable",
    palette=my_colors_E  # tu paleta personalizada
)

plt.xlabel("Estaci√≥n", fontsize=12)
plt.ylabel("Concentraci√≥n de Clorofila (¬µg/L)", fontsize=12)
plt.title("Comparaci√≥n de m√©todos de clorofila por estaci√≥n", fontsize=14)

plt.legend(title="M√©todo", bbox_to_anchor=(1.05, 1), loc='upper left', prop={'size':10})
plt.tight_layout()
plt.show()
```
